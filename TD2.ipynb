```python
import numpy as np
import scipy.integrate as spi
import scipy as sp
```

```python
A=np.array([[-1,-1,-1],[1,0,0],[0,1,0]])
A
```

```python
def sol_hom(t,to):
    return sp.linalg.expm(A*(t-to))

def b(t):
    return(np.array([np.cos(t),0,0]))
```

```python
def second_member(t,t0):
    return spi.quad_vec(lambda s: sol_hom(t,s)@b(s),t,t0)[0]

def solution(t,t0,y0):
    return sol_hom(t,t0)@y0+second_member(t,t0)
```

```python
T=np.linspace(0,20,201)
y0=[0,-1/4,1/2]

res=[solution(t,0,y0)[0] for t in T]
```

```python
def sol(t):
    return (-1/4)*t*np.cos(t) + (1/4)*t*np.sin(t)

real_res=[sol(t) for t in T]

import matplotlib.pyplot as plt
plt.plot(T,res, label = 'approximation')
plt.plot(T,real_res, label = 'real solution')
plt.legend()
plt.show()
```
